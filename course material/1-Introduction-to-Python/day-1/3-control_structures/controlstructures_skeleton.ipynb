```python
# %%
    
#Exercise 3   
LNumbers = [5,23,37,49,50,46,30,46,70]

# 1. while loop
iCount = 0
iSizeofList = len(LNumbers)
i = 0

while i < iSizeofList:
	if LNumbers[i] > 30:
		iCount = iCount + 1
	i = i + 1

print(iCount)

# 2. for loop
iCount = 0

for iNum in LNumbers:
	if iNum > 30:
		iCount = iCount + 1
		
print(iCount)
```
